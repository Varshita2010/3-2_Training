### Noisy Channel MOdel 
- correcting the incorrect words.
- keyword -> noisy channel -> translate/output.
- why we need it?
    - because to check and rectify the errors.
##### Application :-
    1. Spell Checker("Piza" -> "Pizza").
    2. Speech Recognisation("IceCream"-"I Scream").
    3. Machine Translator(french -> english).
- checks logic:-
    - Likelihood
    - Prior probability
## N-gram Language Model:-
- used to estimate the probabilty of a word based on the previous word i.e.,(n-1).
- types of models:-
    1. unigram: (N-1) --> "hi","hello".
    2. bigram: (N-2) --> "Merry" shows "Christmas".
    3. trigram: (N-3) --> "I Love" suggest "Biryani".
- It focus on prediction.
## Evaluation Model:-
- it will evaluate the n-gram language.wheather it works prefectly or not.
- it wil works under the metrics i.e., perplexity.
- lower perplexity with higher model.
- higher perplexity with lower model(confused).
## Smoothing Techniques:-
- distribute mass to unseen n-gram model to avoid zero probablities. or simply andels zero probabilities.
- if it find new word in n-gram return zero as for the new word.
- it has two types: [use this for not getting crashed]
    1. Laplace : ( count - 1)
    2. Add-K: (count - 0.5)
