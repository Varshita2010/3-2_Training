```python
##Project-2
#1-Answer
import re
from nltk.stem import PorterStemmer
data = {
    "A": "I Love this movie!!! #Awesome #friday",
    "B": "I love this movie but the seats were bad",
    "C": "LOVED it! Best Movie Ever!!!"
}
stopwords = {"i", "this", "it", "but", "were", "the"}
stemmer = PorterStemmer()
text = " ".join(data.values())
text = text.lower()
text = re.sub(r'[^a-zA-Z\s]', ' ', text)
tokens = text.split()
tokens = list(filter(lambda w: w not in stopwords, tokens))
tokens = list(map(lambda w: stemmer.stem(w), tokens))
print(tokens)
```
